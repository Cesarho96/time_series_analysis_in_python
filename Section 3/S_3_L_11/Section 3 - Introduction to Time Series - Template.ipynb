### The QQ Plot

- Quantile-Quantile Plot (QQ Plot)
- Used to determine wheter a data set is distributed a certain way.
- Usually showcases how the data fits a Normal Distribution

The QQ-plot Takes all the values a variable can take, and arranges them in accending order. 
The y axis expresses the values with the highest ones at the top and the lowest at the bottom. 
The x axis represents the theoretical quantiles of the dataset (how many standar deviations away from the mean these values are) 
The red diagonal line on the screen represents what the data points should follow if they are normally distributed


```python
import scipy.stats
import pylab
```

```python
scipy.stats.probplot(df_comp.spx, plot = pylab)
plt.title("QQ Plot", size = 24)
pylab.show()
```
