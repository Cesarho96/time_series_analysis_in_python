## Importing the relevant packages


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.graphics.tsaplots as sgt
from statsmodels.tsa.arima.model import ARIMA
from scipy.stats.distributions import chi2
import statsmodels.tsa.stattools as sts 
import seaborn as sns
sns.set()
```

## Importing the Data and Pre-processing 

```python
raw_csv_data = pd.read_csv("./../../Section 3/S_3_L_11/Index2018.csv") 
df_comp=raw_csv_data.copy()
df_comp.date = pd.to_datetime(df_comp.date, dayfirst = True)
df_comp.set_index("date", inplace=True)
df_comp=df_comp.asfreq('b')
df_comp=df_comp.fillna(method='ffill')
```

```python
df_comp['market_value']=df_comp.ftse
```

```python
del df_comp['spx']
del df_comp['dax']
del df_comp['ftse']
del df_comp['nikkei']
size = int(len(df_comp)*0.8)
df, df_test = df_comp.iloc[:size], df_comp.iloc[size:]
```

## LLR Test

```python
def LLR_test(mod_1, mod_2, DF=1):
    L1 = mod_1.fit().llf
    L2 = mod_2.fit().llf
    LR = (2*(L2-L1))
    p = chi2.sf(LR, DF).round(3)
    return p
```

## The DF-Test

```python
sts.adfuller(df.market_value)
```

After running the cell, we see that the P value equals 0.33. so can't conclusively say the data comes from a stationary process

Extracted from a non-stationary process.  
We shouldn't rely on AR models to make accurate forecasts.  
But what can we do in such cases???  
Transforming the data set, so that it fits the "stationary" assumptions.  
In finance, the common approach is to use returns instead of prices when measuring financial indices.

## Using Returns

```python
df['returns'] = df.market_value.pct_change(1).mul(100)
df = df.iloc[1:]
```

```python
sts.adfuller(df.returns)
```

The T statistic is significantly smaller than the 1% critical value of -3.43 which we see below as well. Thus we can confidently state that the data comes from a stationary.

## ACF and PACF for Returns

```python
sgt.plot_acf(df.returns, lags=40, zero = False)
plt.title("ACF FTSE Returns", size=24)
plt.ylim(-0.1, 0.1)
plt.show()
```

```python
sgt.plot_pacf(df.returns, lags = 40, zero = False, method = ('ols'))
plt.title("PACF FTSE Returns", size=24)
plt.ylim(-0.1, 0.1)
plt.show()
```

## AR(1) for Returns

```python
model_ret_ar_1 = ARIMA(df.returns, order = (1,0,0))
```

```python
results_ret_ar_1 = model_ret_ar_1.fit()
```

```python
results_ret_ar_1.summary()
```

## Higher-Lag AR Models for Returns

```python
model_ret_ar_2 = ARIMA(df.returns, order = (2,0,0))
results_ret_ar_2 = model_ret_ar_2.fit()
results_ret_ar_2.summary()
```

```python
LLR_test(model_ret_ar_1, model_ret_ar_2)
```

## Normalizing Values

```python
benchmark = df.market_value.iloc[0]
```

```python
df['norm'] = df.market_value.div(benchmark).mul(100)
```

```python
sts.adfuller(df.norm)
```

```python
bench_ret = df.returns.iloc[0]
df['norm_ret'] = df.returns.div(bench_ret).mul(100)
sts.adfuller(df.norm_ret)
```
